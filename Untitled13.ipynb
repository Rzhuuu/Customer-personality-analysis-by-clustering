```python
import numpy as np
import pandas as pd
import sklearn # ML library
from sklearn import metrics #score functons, performance metrics, pairwise metrics and distance computations
from sklearn.preprocessing import LabelEncoder # Encode target labels between 0 and n_classes-1
from sklearn.preprocessing import StandardScaler # Standardize features by removing the mean and scaling to unit variance
from sklearn.decomposition import PCA
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans # Kmeans clustering
from sklearn.cluster import AgglomerativeClustering
import datetime
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import colors
from matplotlib.colors import ListedColormap
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D # generating 3D plots using the mplot3d toolkit
import warnings
import sys # system specific parameters and functions
if not sys.warnoptions:
    warnings.simplefilter("ignore") # overriding the default filter
np.random.seed(42)
```

```python
data = pd.read_csv('/content/marketing_campaign.csv', sep="\t")
df = pd.DataFrame(data)
rows = len(df.axes[0])
cols = len(df.axes[1])
print('number of rows:', rows)
print('number of columns:', cols)
df.head()
```

```python
df.info()
```

```python
df = df.dropna()
print('The number of rows after dropping rows with missing values is:', len(df.axes[0]))
```

```python
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"], format = 'mixed')
dates = []
for i in df["Dt_Customer"]:
  i = i.date() # Extract the date portion(year, month, day)
  dates.append(i)
print("The latest customer's enrolment date in the records:", max(dates))
print("The oldest customer's enrolment date in the records:", min(dates))
```

```python
days=[]
d1=max(dates)
for i in dates:
  delta = d1 - i
  days.append(delta)
df["Customer_For"] = days
df["Customer_For"] = pd.to_numeric(df['Customer_For'], errors='coerce') # if there is an error encountered during the conversion of strings to datetime objects, pandas will set the problematic values to NaT (Not a Time), treating them as missing values.
df["Customer_For"]
```

```python
print("Total categories in the feature Marital_Status:\n", df["Marital_Status"].value_counts(),'\n')
print("Total categories in the feature Education:\n", df['Education'].value_counts(), '\n')
```

```python
# Feature Engineering
df['Age'] = 2024 - df['Year_Birth'] # customer age today
df['Total_Spent'] = df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntSweetProducts'] + df['MntGoldProds']
df["Living_With"] = df["Marital_Status"].replace({"Married":"Partner", "Together":"Partner", "Absurd":"Alone","Widow":"Alone","YOLO":"Alone","Divorced":"Alone", "Single":"Alone"})
df['Children'] = df['Teenhome'] + df['Kidhome']
df['Family_Size'] = df['Living_With'].replace({'Alone':1, 'Partner': 2})+df['Children']
df['Is_Parent'] = np.where(df.Children>0,1,0)
df['Education'] = df['Education'].replace({'Basic':'Undergraduate', '2n Cycle': 'Undergraduate', 'Graduation':'Graduate','Master':'Postgraduate', 'PhD':'Postgraduate'})
df = df.rename(columns = {'MntWines':'Wines', 'MntFruits':'Fruits', 'MntMeatProducts':'Meat', 'MntFishProducts':'Fish','MntSweetProducts':'Sweets', 'MntGoldProds':'Gold'})
columns_to_drop = ['Marital_Status', 'Dt_Customer', 'Z_CostContact', 'Z_Revenue', 'Year_Birth', 'ID']
df = df.drop(columns = columns_to_drop, axis=1)
```

```python
df.describe()
```

```python
sns.set(rc={'axes.facecolor':'white', 'figure.facecolor':'white'})
palette = ["#84d1e8", "#84d1e8", "#c281e3", "#B9C0C9", "#9F8A78", "#F3AB60"]
cmap = colors.ListedColormap(palette)
To_plot = ['Income','Recency', 'Customer_For', 'Total_Spent', 'Age', 'Is_Parent']
print('Relative plot of selected features:')
plt.figure()
sns.pairplot(df[To_plot], hue = 'Is_Parent', palette = ["#84d1e8", "#9E726F"]) # hue to color plot aspects based on the values of a specific variable.
plt.show()
```

```python
df = df[df['Age']<90]
df = df[df['Income']<600000]
print('The total number of data points after dropping outliers are:', len(df))
```

```python
categorical_columns = df.select_dtypes(include=['object']).columns
numerical_data = df.drop(columns=categorical_columns)
correlation_matrix = numerical_data.corr()
plt.figure(figsize=(20,20))
sns.heatmap(correlation_matrix,annot=True,cmap=cmap,center=0) # cmap is colormap
```

```python
LE = LabelEncoder()
for i in categorical_columns:
    df[i] = df[[i]].apply(LE.fit_transform)
print ('All features are now numerical')
```

```python
data = df.copy()
cols_del = ['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1', 'AcceptedCmp2', 'Complain', 'Response']
data = data.drop(cols_del, axis=1)
scaler = StandardScaler()
scaler.fit(data)
scaled_data = pd.DataFrame(scaler.transform(data),columns=data.columns)
print('All features have been scaled')
```

```python
print('Dataframe to be used for further modeling:')
scaled_data.head()
```

```python
# Dimensionality reduction
# PCA
pca = PCA(n_components =3)
pca.fit(scaled_data)
PCA_data = pd.DataFrame(pca.transform(scaled_data), columns = (['col1', 'col2', 'col3']))
PCA_data.describe().T
```

```python
# 3D projection
x = PCA_data['col1']
y = PCA_data['col2']
z = PCA_data['col3']
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection = '3d') # only one row, one column, and this subplot is the first one.
ax.scatter(x,y,z, c='#e3bf81', marker = 'o') # c means color.  marker style being o means circular marker.
ax.set_title('A 3D projection of data in reduced dimension')
plt.show()
```

```python
# Agglomerative is bottom up approach. Divisive is starting at top and divide into multiple.
print ('Elbow method to determine trhe number of clusters to be formed:')
Elbow_M = KElbowVisualizer(KMeans(), k=10)
Elbow_M.fit(PCA_data)
Elbow_M.show()
```

```python
AC = AgglomerativeClustering(n_clusters = 4)
yhat_AC = AC.fit_predict(PCA_data)
PCA_data['Clusters'] = yhat_AC
df['Clusters'] = yhat_AC
```

```python
fig = plt.figure(figsize=(10,8))
ax = plt.subplot(111, projection='3d', label ='bla')
ax.scatter(x,y,z, s = 40, c = PCA_data['Clusters'], marker = 'o', cmap = cmap)# s is the size of the markers. c specifies the color of the markers based on the cluster.
ax.set_title('Plot of clusters')
plt.show()
```

```python
pal = ['#b8c748', '#B9C0C9', '#53b6e0', '#F3AB60']
pl = sns.countplot(x = df['Clusters'], palette = pal)
pl.set_title('Number of customers in each cluster')
plt.show()
```

```python
pl = sns.scatterplot(data = df, x=df['Total_Spent'], y=df['Income'], hue=df['Clusters'], palette = pal)
pl.set_title("Cluster's profile based on income and spending")
plt.legend()
plt.show()
```

```python
plt.figure()
pl = sns.swarmplot(x = df['Clusters'], y = df['Total_Spent'], color= '#CBEDDD', alpha = 0.5)
pl = sns.boxenplot(x = df['Clusters'], y = df['Total_Spent'], palette = pal)
plt.show()
```
